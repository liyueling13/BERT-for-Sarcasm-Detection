```python
# lowercase comments
sarc_test['comment'] = sarc_test['comment'].apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# remove non-words

import re
sarc_test['comment'] = sarc_test['comment'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x) if isinstance(x, str) else x)
```

```python
def remove_one_letter(text):
  words = text.split()
  filtered_words = [word for word in words if len(word)>1]
  return ' '.join(filtered_words)

sarc_test['comment'] = sarc_test['comment'].apply(remove_one_letter)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

sarc_test['comment'] = sarc_test['comment'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]) if isinstance(x, str) else x)
```

```python
sarc_test['comment'] = sarc_test['comment'].apply(lambda x: ' '.join(word for word in x.split() if word.lower() not in stop_words))

# removed stopwords
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
import scattertext as st
```

```python
# let's build a scattertext corpus
# with sarc_sm excerpted from sarc_test

sarc_corpus = st.CorpusFromPandas(sarc_sm,
                             category_col='label',
                             text_col='comment',
                             nlp=nlp).build()
```

```python
# these terms distinguish the test corpus from the english language corpus

sarc_corpus.get_scaled_f_scores_vs_background().index[:10]
```

```python
# distinguishing words in non-sarcastic comments

term_freq_df = sarc_corpus.get_term_freq_df()
term_freq_df['Not Sarcastic Score'] = sarc_corpus.get_scaled_f_scores('0')
term_freq_df.sort_values(by='Not Sarcastic Score', ascending=False).index[:10]
```

```python
# distinguishing words in sarcastic comments

term_freq_df = sarc_corpus.get_term_freq_df()
term_freq_df['Sarcastic Score'] = sarc_corpus.get_scaled_f_scores('1')
term_freq_df.sort_values(by='Sarcastic Score', ascending=False).index[:10]
```

```python
from scattertext import SampleCorpora, PhraseMachinePhrases, dense_rank, RankDifference, AssociationCompactor, produce_scattertext_explorer
from scattertext.CorpusFromPandas import CorpusFromPandas

html = st.produce_scattertext_explorer(sarc_corpus,
                                       category='0',
                                       category_name='Non-Sarcastic Comments',
                                       not_category_name='Sarcastic Comments',
                                       width_in_pixels=1000)


# Open the html for interactive visualization
open("Sarcasm Scattertext Explorer.html", 'wb').write(html.encode('utf-8'))
```
